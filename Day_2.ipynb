While the Red-Nosed Reindeer nuclear fusion/fission plant appears to contain no sign of the Chief Historian, the engineers there run up to you as soon as they see you. Apparently, they still talk about the time Rudolph was saved through molecular synthesis from a single electron.

You turn to check if The Historians are waiting for you, but they seem to have already divided into groups that are currently searching every corner of the facility. You offer to help with the unusual data.

The unusual data (your puzzle input) consists of many reports, one report per line. Each report is a list of numbers called levels that are separated by spaces. For example:

```
7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9
```
This example data contains six reports each containing five levels.

The engineers are trying to figure out which reports are safe. The Red-Nosed reactor safety systems can only tolerate levels that are either gradually increasing or gradually decreasing. So, a report only counts as safe if both of the following are true:

    The levels are either all increasing or all decreasing.
    Any two adjacent levels differ by at least one and at most three.

In the example above, the reports can be found safe or unsafe by checking those rules:
```
    7 6 4 2 1: Safe because the levels are all decreasing by 1 or 2.
    1 2 7 8 9: Unsafe because 2 7 is an increase of 5.
    9 7 6 2 1: Unsafe because 6 2 is a decrease of 4.
    1 3 2 4 5: Unsafe because 1 3 is increasing but 3 2 is decreasing.
    8 6 4 4 1: Unsafe because 4 4 is neither an increase or a decrease.
    1 3 6 7 9: Safe because the levels are all increasing by 1, 2, or 3.
```
So, in this example, 2 reports are safe.

Analyze the unusual data from the engineers. How many reports are safe?

````python
test_input = """7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9"""


def test_function(f):
    assert f(test_input) == 2


def test_line_function(f):
    assert f([7, 6, 4, 2, 1]) == True
    assert f([1, 2, 7, 8, 9]) == False
    assert f([9, 7, 6, 2, 1]) == False
    assert f([1, 3, 2, 4, 5]) == False
    assert f([8, 6, 4, 4, 1]) == False
    assert f([1, 3, 6, 7, 9]) == True


def sign(i):
    if i < 0:
        return -1
    if i > 0:
        return 1
    else:
        return 0


def is_line_safe(line):
    if len(line) <= 1:
        return False
    direction = sign(line[0] - line[1])
    if direction == 0:
        return 0
    # print('line: ', line)
    # print('\tdirection: ', direction)
    # print('\tline_len', len(line))
    a = [
        x - y
        for x, y in zip(line[:-1], line[1:])
        if (abs(x - y) <= 3) and (sign(x - y) == direction)
    ]

    # print('\ta_len',len(a))
    # print('\ta:',a)
    return len(a) == len(line) - 1


test_line_function(is_line_safe)


def are_levels_safe(text_input):
    lines = [[int(j) for j in i.split(" ") if j != ""] for i in text_input.split("\n")]

    return sum([is_line_safe(l) for l in lines])


test_function(are_levels_safe)
````

````python
test_input = """7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9"""


def test_function(f):
    assert f(test_input) == 4


def test_line_function(f):
    assert f([7, 6, 4, 2, 1]) == True
    assert f([1, 2, 7, 8, 9]) == False
    assert f([9, 7, 6, 2, 1]) == False
    assert f([1, 3, 2, 4, 5]) == True
    assert f([8, 6, 4, 4, 1]) == True
    assert f([1, 3, 6, 7, 9]) == True


def validate_line(line):
    if len(line) <= 1:
        return False
    direction = sign(line[0] - line[1])

    if direction == 0:
        return 0
    # print("line: ", line)
    # print("\tdirection: ", direction)
    # print("\tline_len", len(line))
    a = [
        x - y
        for x, y in zip(line[:-1], line[1:])
        if (abs(x - y) <= 3) and (sign(x - y) == direction)
    ]

    # print("\ta_len", len(a))
    # print("\ta:", a)
    return len(a) >= len(line) - 1


def is_line_safe(line):

    to_check = []
    to_check.append(line)
    for i in range(len(line)):
        to_check.append(line[:i] + line[i + 1 :])

    return sum([validate_line(i) for i in to_check]) > 0


test_line_function(is_line_safe)


def are_levels_safe(text_input):
    lines = [[int(j) for j in i.split(" ") if j != ""] for i in text_input.split("\n")]

    return sum([is_line_safe(l) for l in lines])


test_function(are_levels_safe)
````
